```python
import numpy as np
from matplotlib import pyplot as plt
import base_psx.grid_bubble, base_psx.base
import importlib
importlib.reload(base_psx.grid_bubble)
import qiskit
import types

four_term_psr = {
    'alpha': np.pi / 2,
    'beta' : 3 * np.pi / 2,
    'd_plus' : (np.sqrt(2) + 1) / (4*np.sqrt(2)),
    'd_minus': (np.sqrt(2) - 1) / (4*np.sqrt(2))
}

num_shots = 10000
learning_rate = 0.04
noise_prob = 0.00
backend = qiskit.Aer.get_backend('qasm_simulator')

def measure(qc: qiskit.QuantumCircuit, qubits, cbits=[]):
    """As its function name
    Args:
        qc (qiskit.QuantumCircuit): measuremed circuit
        qubits (np.ndarray): list of measuremed qubit
        cbits (list, optional): classical bits. Defaults to [].
    Returns:
        qiskit.QuantumCircuit: added measure gates circuit
    """
    if cbits == []:
        cbits = qubits.copy()
    for i in range(0, len(qubits)):
        qc.measure(qubits[i], cbits[i])
    counts = qiskit.execute(
            qc, backend=backend,
            shots=num_shots).result().get_counts()
    return counts.get("0" * len(qubits), 0) / num_shots

def grad_loss(qc: qiskit.QuantumCircuit, create_circuit_func: types.FunctionType,
              thetas: np.ndarray, **kwargs):
    """Return the gradient of the loss function
    L = 1 - |<psi~|psi>|^2 = 1 - P_0
    => nabla_L = - nabla_P_0 = - r (P_0(+s) - P_0(-s))
    Args:
        - qc (QuantumCircuit): The quantum circuit want to calculate the gradient
        - create_circuit_func (Function): The creating circuit function
        - thetas (np.ndarray): Parameters
        - c_0 (float): cost value
        - **kwargs: additional parameters for different create_circuit_func()
    Returns:
        - np.ndarray: the gradient vector
    """
    for i in range(0, len(thetas)):
        thetas1, thetas2 = thetas.copy(), thetas.copy()
        thetas3, thetas4 = thetas.copy(), thetas.copy()
        thetas1[i] += four_term_psr['alpha']
        thetas2[i] -= four_term_psr['alpha']
        thetas3[i] += four_term_psr['beta']
        thetas4[i] -= four_term_psr['beta']
        qc1 = create_circuit_func(qc.copy(), thetas1, **kwargs)
        qc2 = create_circuit_func(qc.copy(), thetas2, **kwargs)
        qc3 = create_circuit_func(qc.copy(), thetas3, **kwargs)
        qc4 = create_circuit_func(qc.copy(), thetas4, **kwargs)
        grad_loss[i] = - (four_term_psr['d_plus'] * (
            measure(qc1, list(range(qc1.num_qubits))) -
            measure(qc2, list(range(qc2.num_qubits)))) -four_term_psr['d_minus'] * (
            measure(qc3, list(range(qc3.num_qubits))) -
            measure(qc4, list(range(qc4.num_qubits)))))
    return grad_loss
    

def create_fully_cry(qc: qiskit.QuantumCircuit, thetas):
    for i in range(0, qc.num_qubits):
        if i == qc.num_qubits - 1:
            qc.cry(thetas[i], qc.num_qubits - 1, 0)
        else:
            qc.cry(thetas[i], 0, i + 1)
    return qc

qc = qiskit.QuantumCircuit(3, 3)
thetas = np.ones(3)
print(grad_loss(qc, create_fully_cry, thetas))
```
